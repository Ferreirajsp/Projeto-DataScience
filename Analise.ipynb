## Carregando o conjunto de dados 

```python
import pandas as pd

had = pd.read_excel('datascience/HAD_BRASIL_REGIOES_UFS_2024.xlsx', skiprows=8)
```

```python
had
```

```python
had.info()
```

## Preparando os dados para a análise 

**Obs:** Todas as colunas estão sendo identificadas como Object

Antes de resolver esse problema, selecionei algumas colunas que são relevantes para termos uma visão "por cima" das médias das horas-aulas de cada categoria da escola (Federal, Privada, Estadual e Municipal) por seu Estado e Região.


```python
had = had.iloc[:,[1,2,3,4,7,20]]
had
```

Dicionário:
-ug: Unidade Geográfica
-loc: Localização
-depA: Dependência Administrativa
-tei: Total Ensino Infantil
-tef: Total Ensino Fundamental
-tem: Total Ensino Médio

```python
colunas = ["ug", "loc", "depA", "tei", "tef", "tem"]
had.columns = colunas 
```

```python
had
```

```python
df = had.replace('--',0)
```

```python
had.info()
```

## Hora-Aula dos Estados e suas Médias

```python
df.groupby("ug")[["tei", "tef", "tem"]].median().round(2)
```

## Relação de Horas-Aula da Área Urbana e Área Rural

```python
df.groupby("loc")[["tei", "tef", "tem"]].median()
```

## Hora-Aula Estado/Localização

```python
df.groupby(["ug", "loc"])[["tei", "tef", "tem"]].median().round(2)
```

## Resumo Geral dos Dados Inseridos e Análisados

```python
pd.pivot_table(df, values= "tef", index= "ug", columns="depA", aggfunc="median")
```

## Visualização dos Dados
Média Hora-Aula por Dependência Administrativa

```python
import matplotlib.pyplot as plt

df.groupby("depA")[["tei", "tef", "tem"]].median().plot(kind="barh", figsize=(10, 5))
plt.title("Média de Hora-Aula por Dependência Escolar")
plt.xlabel("Quantidade média de Hora-Aula")
plt.ylabel("Tipo de administração")
plt.tight_layout()
plt.show()
```

Média Hora-Aula por Localização

```python
df.groupby("loc")[["tei", "tef", "tem"]].mean().plot(kind="bar", figsize=(8, 5))
plt.title("Média de Hora-Aula por Localização")
plt.ylabel("Média de Hora-Aula")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

Média com Gráfico de Pizza


```python
df.groupby("depA")["tei"].median().plot.pie(autopct="%1.1f%%", figsize=(6, 6), ylabel="")
plt.title("Distribuição da Dependência Administrativa para o total do Ensino Infantil")
plt.show()
```

```python
df.groupby("depA")["tef"].median().plot.pie(autopct="%1.1f%%", figsize=(6, 6), ylabel="")
plt.title("Distribuição da Dependência Administrativa para o total do Ensino Fundamental")
plt.show()
```

```python
df.groupby("depA")["tem"].median().plot.pie(autopct="%1.1f%%", figsize=(6, 6), ylabel="")
plt.title("Distribuição da Dependência Administrativa para o total do Ensimo Médio")
plt.show()
```
